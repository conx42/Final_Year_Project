```python
import nltk
with open('C:/Users/91629/OneDrive/Desktop/7th_SEM_proj_NLTK/custom_Input.c', 'r') as file:
    c_code = file.read()
tokens=nltk.word_tokenize(c_code)
n=len(tokens)
print('After tokenization of the External C file:\n')
print(tokens,'\n')
maxval=2**31-1
minval=-2**31
outer_conditional_statements={}
nested_conditional_statements=[] # <-- inner_conditional_statements=[]


c_operators = ['>','<','==', '!=', '<=', '>=']


def eval_cond(x, count):
    global CondCounter
    if(count==1 and x=='('):
        CondCounter=0
        count+=1
    if x =='(':
        CondCounter+=1
    elif x==')':
        CondCounter-=1
    elif CondCounter==0:
        return True
    return False

# function for tracking outer & inner if condition together
def outer_inner_condition_curlyBraces(token,i):
    global track_curly_brackets_for_inner_nd_outer_if_conditions
    if i==0:
        track_curly_brackets_for_inner_nd_outer_if_conditions=i
    if token == '{':
        track_curly_brackets_for_inner_nd_outer_if_conditions+=1
    elif token == '}':
        track_curly_brackets_for_inner_nd_outer_if_conditions-=1
    if track_curly_brackets_for_inner_nd_outer_if_conditions == 1:
        return 'outer'
    elif track_curly_brackets_for_inner_nd_outer_if_conditions > 1:
        return 'inner'
    # else:
    #     print('track_curly_brackets_for_inner_nd_outer_if_conditions= ',track_curly_brackets_for_inner_nd_outer_if_conditions)
    #     print('So, Program ends')

# outer-IF_er-Vitor_total_innerConditions or also khonw as(aka) gass
global gas
gas=0

# Detect outer and Inner if together
for i in range(0,n):
    take_condition=outer_inner_condition_curlyBraces(tokens[i],i)
    # for outer if
    if tokens[i] == 'if' and take_condition=='outer':
        # add this 'if' condition to outer if
        gas+=1
        outer_conditional_statements[gas]=[]
        for j in range(i+1,n):
            if eval_cond(tokens[j], 1)== False:
                outer_conditional_statements[gas].append(tokens[j])
                # print('token: ',tokens[j])
                # print('condition: ',Condition)
            else:
#                 print(eval_cond(tokens[j]))
                break
    elif tokens[i]=='if' and take_condition=='inner':
        for j in range(i+1,n):
            if eval_cond(tokens[j], 1)== False:
                nested_conditional_statements.append(tokens[j])
                # print('token: ',tokens[j])
                # print('condition: ',Condition)
            else:
#                 print(eval_cond(tokens[j]))
                break

print("\nnested_conditional_statements: ",nested_conditional_statements,"\n")
print("\nouter_conditional_statements: ", outer_conditional_statements,"\n")

# global gasIndex
# gasIndex=0
# function for tracking outer & inner if condition together
# _track=0
def _tracker(token, idx):
    global _track
    if idx==0:
        _track=0
    if token == '{':
        _track+=1
    elif token == '}':
        _track-=1
    if _track == 1:
        return 'outer'
    elif _track > 1:
        return 'inner'
    else:
        return 'null'

printInner=0
    
def printNesTed_conditions(Index):
    while (Index<n):
        con = _tracker(tokens[Index], Index)
        if tokens[Index]=='if' and con=='inner':
#             Index+=1
            if printInner==0:   
                while(printInner<(printInner+4) and printInner<n):
                    print(f"\t\t {nested_conditional_statements[printInner]} \n")
                    printInner+=1
            else:
                while(printInner<(printInner+5) and printInner<n):
                    print(f"\t\t {nested_conditional_statements[printInner]} \n")
                    printInner+=1
        Index=Index+1
    return Index
        

def Access_ConditionalStatements_Beautify():
    gasIndex=0
#     printList
    for key,value in outer_conditional_statements.items(): 
        if key==2:
            print(f"{key}nd - Outer Conditional Statement -> {value}")
            print(f"\tUnder {key}nd - conditional statement - NESTED Conditional statements are - ")
            gasIndex=printNesTed_conditions(gasIndex)
        elif key==3:
            print(f"{key}rd - Outer Conditional Statement -> {value}")
            print(f"\tUnder {key}rd - conditional statement - NESTED Conditional statements are - ")
            gasIndex=printNesTed_conditions(gasIndex)
        else:
            print(f"{key}st - Outer Conditional Statement -> {value}")
            print(f"\tUnder {key}st - conditional statement - NESTED Conditional statements are - ")
            gasIndex=printNesTed_conditions(gasIndex)
        


Access_ConditionalStatements_Beautify()
```
